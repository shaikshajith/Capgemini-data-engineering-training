```python
%sql
---1
select *,
case when experience>=8 and performance_rating="A" then salary*0.20
     when experience>=5 and performance_rating="B" then salary*0.15
     when performance_rating="C" then salary*0
     else salary+salary*0.10 end as New_salary
     from Employ;
```

```python
%sql
--2
select *,
case when performance_rating="A" and department="Finance" then salary*0.20
    when performance_rating="B" and department="Finance" then salary*0.15
    when performance_rating="C" and department="Finance" then salary*0.05
    when performance_rating="A" and department="Engineering" then salary*0.18
    when performance_rating="B" and department="Engineering" then salary*0.12
    when performance_rating="C" and department="Engineering" then salary*0.03
    else salary*0.10 end as Salary_Hike from Employ
```

```python
%sql
--3
select *,
case when salary>=80000 and Performance_rating="A" then "High Performer"
when salary between 50000 and 80000 and performance_rating="B" then "Mid Performer"
when salary<=50000 and performance_rating="C" then "Low Performer"
else "Poor Performer"
end as Rating
from Employ
```

```python
%sql
--2
select *,case  when department="Finance" then
                case when performance_rating="A" and experience>=10 then salary*0.25
                else salary*0.20
                end 
      when department="HR" then
                case when performance_rating="B" and experience>=5 then salary*0.15
                else salary*0.10
                end 
else salary*0.08 end as Salary_new
from Employ; 
```

```python
%sql
--1
select * ,
case when performance_rating="A" then 
    case when salary>80000 and experience>5 then salary+salary*0.25
    when salary between 50000 and 80000  then salary+salary*0.15
    else salary+salary*0.20
    end
     
when performance_rating="B" then 
      case when experience>5 then salary+salary*0.12
      else salary+salary*0.10
      end
when performance_rating="C" then salary 
end as New_salary
from Employ;
```
