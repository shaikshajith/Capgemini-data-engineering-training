```python
%sql
CREATE TABLE departments (
dept_id INT PRIMARY KEY,
dept_name VARCHAR(50)
);

INSERT INTO departments (dept_id, dept_name) VALUES
(1, 'HR'),
(2, 'IT'),
(3, 'Finance'),
(4, 'Marketing'),
(5, 'Sales');
```

```python
%sql
CREATE TABLE projects (
project_id INT PRIMARY KEY,
project_name VARCHAR(50),
emp_id INT
);

INSERT INTO projects (project_id, project_name, emp_id) VALUES
(1, 'Project A', 1),
(2, 'Project B', 2),
(3, 'Project C', 3),
(4, 'Project D', 4),
(5, 'Project E', 5);
```

```python
%sql
select e1.emp_name as Employee_name,e2.emp_name as Manager_name
from employees e1 join employees e2 on e1.manager_id = e2.emp_id ; 
```

```python
%sql
select e.emp_name , d.dept_name
from employees e left join departments d on e.dept_id=d.dept_id;
```

```python
%sql
select *
from employees e left join departments d on e.dept_id = d.dept_id where e.dept_id =null;
```

```python
%sql
select e.emp_name,p.project_name
from employees e left join projects p on e.emp_id=p.emp_id
```
